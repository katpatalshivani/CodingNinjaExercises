# Power function

```python
def power(x, n):
    # Please add your code here
    if n==0:
        return 1
    
    ans = power(x,n/2)
    
    if n%2 == 0:
        return ans*ans
    
    else:
        return ans*ans*x

# Main
from sys import setrecursionlimit
setrecursionlimit(11000)
x, n=list(int(i) for i in input().strip().split(' '))
print(power(x, n))
```
