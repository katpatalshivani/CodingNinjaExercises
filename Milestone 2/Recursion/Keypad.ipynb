```python
def getstring(d):
    if d ==2:
        return "abc"
        
    if d == 3:
        return "def"
        
    if d==4:
        return "ghi"
    
    if d==5:
        return "jkl"
        
    if d==6:
        return "mno"
        
    if d==7:
        return "pqrs"
    
    if d==8:
        return "tuv"
        
    if d==9:
        return "wxyz"



def keypad(n):
    if n == 0:
        output=[]
        output.append("")
        return output
    
    smallerInput = n // 10
    digit = n % 10
    
    smallAns = keypad(smallerInput)
    option = getstring(digit)
    
    o = []  
    for c in option:
        for i in smallAns:
            ans = i+c
            o.append(ans)
    return o

n = 234
finalOp = keypad(n)
print(len(finalOp))
for a in finalOp:
    print(a)
        
    
    
```
