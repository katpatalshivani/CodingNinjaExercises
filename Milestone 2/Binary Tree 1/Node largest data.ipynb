```python
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def largestData(root):
	# Your code goes here
    if root == None:
        return -1
    
    left_large=largestData(root.left)
    right_large=largestData(root.right)
    
    largest=max(root.data,left_large,right_large)
    
    return largest
    
    
    
    
    

    

def treeInput():
    rootData=int(input())
    
    if rootData ==-1:
        return None
    
    
    root=BinaryTreeNode(rootData)
    
    leftTree = treeInput()
    rightTree=treeInput()
    root.left=leftTree
    root.right=rightTree
    
    return root

def printTree(root):
    if root == None:
        return
    
    print(root.data,end=":")
    
    if root.left !=None:
        print("L ",root.left.data,end=",")
        
    if root.right !=None:
        print("R ",root.right.data,end=" ")
    print()
        
    printTree(root.left)
    printTree(root.right)
    
    
root=treeInput()
printTree(root)
print(largestData(root))
        
    
    
        
    
```
