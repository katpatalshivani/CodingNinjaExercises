```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
        
        
def changeToDepthTree(root,d=0) :
	#Your code goes here
    
    if root==None:
        return
    
    root.data=d
    
    changeToDepthTree(root.left,d+1)
    changeToDepthTree(root.right,d+1)
    
def printInorder(root):
    if root == None:
        return
    
    print(root.left.data, end= " ")
    print(root.data,end=" ")
    print(root.right.data,end=" ")
    
        
def inputTree():
    rootdata=int(input())
    
    if rootdata==-1:
        return
    
    root=BinaryTreeNode(rootdata)
    
    leftTree = inputTree()
    rightTree=inputTree()
    root.left=leftTree
    root.right=rightTree
        
    return root
        
def printTree(root):
    if root == None:
        return None
    
    print(root.data,end=":")
    
    if root.left !=None:
        print(root.left.data, end=",")
        
    if root.right !=None:
        print(root.right.data, end=" ")
        
    print()
    printTree(root.left)
    printTree(root.right)
        
root=inputTree()
printTree(root)
changeToDepthTree(root)
printInorder(root)
```
