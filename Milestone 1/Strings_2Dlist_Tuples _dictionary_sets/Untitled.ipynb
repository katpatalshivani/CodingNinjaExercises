```python
n=int(input())
li=[int(x) for x in input().split()]


for i in range(0,n):
    
    for j in range(i+1,n):
        #print(j)
        if li[i]<=li[j]:
            break
    if j==n-1:
        print(li[i],end=' ')
            
#print(leader, " ",num)


    
```

```python
n=int(input())
l=list(map(int,input().split()))
ma=l[-1]
l=l[::-1]
k=[l[0]]
for i in range(1,n):
    if l[i]>=ma:
        ma=l[i]
        k.append(l[i])
k=k[::-1]
print(*k)
```
