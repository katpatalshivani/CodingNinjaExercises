```python
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None


def findNodeRec(head, n) :
	#Your code goes here
    if head is None or head.next is None:
        return -1
    
    if head.data==n:
        return 0
    
    x= findNodeRec(head.next,n)
    
    if x==-1:
        return x
    
    
    return x+1






























def takeInput() :
    head = None
    tail = None

    datas = [int(x) for x in input().split() ]

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head




def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main

head = takeInput()
n = int(input())    

print(findNodeRec(head, n))

    
```
