```python
# Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None

'''
1->2->3->4->5->N
'''
def lenLL(head):
    
    count =0
    
    while head is not None:
        count=count+1
        head=head.next
        
    return count


def removeDuplicates(head) :
    '''
    if lenLL(head)== 0:
        return head
    
    else:
        t1=head
        t2=head
        
        while t2 is not None:
            while t1 != t2:
                t2=t2.next
            t1.next=t2
            t1=t2
            
    return head
    '''
    
    
    
    
   
    
    
#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = [int(x) for x in input().split()]

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head


#to print the linked list 
def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()




head = takeInput()
#n = int(input())

head = removeDuplicates(head)
printLinkedList(head)

    
```
