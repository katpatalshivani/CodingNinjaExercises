# optimised solution of complexity O(n)

```python
class Node:
    def __init__(self,data):
        self.data= data
        self.next = None
        
    def printLL(head):
        while head is not None:
            print(str(head.data) + "->",end="")
            head=head.next
        print("None")
        
        
    def takeInput():
        inputList=[int(ele) for ele in input().split()]
        
        head=None
        tail=None
        
        while currData in inputList:
            if currData==-1:
                break
                
            newNode= Node(currData)
            
            if head is None:
                head = newNode
                tail = newNode
                
            else:
                tail=newNode
                tail.next = newNode
                
        
        return head
    
head=takeInput()
printLL(head)
```
