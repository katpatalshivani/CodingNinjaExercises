```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    
```

```python
import queue
def takeInputLevelwise():
    
    q=queue.Queue()
    
    print ("Enter root")
    rootData = int(input())
    
    if rootData is None:
        return
    
    root = BinaryTreeNode(rootData)
    
    q.put(root)
    
    while not(q.empty()):
        curr=q.get()
        
        
        print("enter left of ", curr.data)
        leftChild = int(input())
            
        if leftChild != -1:
            leftNode = BinaryTreeNode(leftChild)
            curr.left = leftNode
            q.put(leftNode)
                
                
        print("enter right of ", curr.data)
        rightChild = int(input())
             
        if rightChild != -1:
            rightNode = BinaryTreeNode(rightChild)
            curr.right = rightNode
            q.put(rightNode)
                
    return root
                
            
```

```python
def findPathBST(root,data):
    if root == None:
        return None
    
    if root == data:
        lb=list()
        lb.append(root.data)
        return lb
    
    leftOutput=findPathBST(root.left,data)
    if leftOutput!= None:
        lb.append(root.data)
        return leftOutput
    
    rightOutput=findPathBST(root.left,data)
    if rightOutput!=None:
        lb.append(root.data)
        return rightOutput
    
    else:
        return None
```

```python
import queue
def printTree(root):
       
    q=queue.Queue()
    q.put(root)
    
    while not(q.empty()):
        curr=q.get()
        
        
        print(curr.data, end = ":")
            
        if curr.left != None:
            q.put(curr.left)
            print("L->", curr.left.data,end=",")
            
        else:
            print("L->", -1,end=",")
            
            
        if curr.right != None:
            q.put(curr.right)
            print("R->",curr.right.data)
        else:
            print("R->", -1)
            
            
    
```

```python
root=takeInputLevelwise()
printTree(root)
print(findPathBST(root,3))
```
