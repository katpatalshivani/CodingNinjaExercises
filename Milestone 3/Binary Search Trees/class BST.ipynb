```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right= None
```

```python
class BST:
    def __init__(self):
        self.root=None
        self.numNode = 0
       #printing function
    
    def printTreeHelper(self,root):
        if root == None:
            return
        
        print(root.data,end=":")
        
        if root.left!=None:
            print("L->",root.left.data,end=",")
            
        if root.right!=None:
            print("R->",root.right.data,end=" ")
        
        print()
        self.printTreeHelper(root.left)
        self.printTreeHelper(root.right)

        
    def printTree(self):
        self.printTreeHelper(self.root)
    
    #is node present function
    def isPresentHelper(self,root,data):
        if root==None:
            return False
        
        if root.data == data:
            return True
        
        if root.data> data:
            return self.isPresentHelper(root.left,data)
        else:
            return self.isPresentHelper(root.right,data)
            
    def isPresent(self,data):
        return self.isPresentHelper(self.root,data)
    
    #insert function
    def insertHelper(self,root,data):
        if root==None:
            node = BinaryTreeNode(data)
            return node
        
        if root.data>data:
            root.left = self.insertHelper(root.left,data)
            return root
        
        else:
            root.right=self.insertHelper(root.right,data)
            return root
    
    def insert(self,data):
        self.numNode +=1
        self.root=self.insertHelper(self.root,data)
        
    def min(self,data):
        if root == None:
            return 100000
        
        if root.left == None:
            return root.data
        
        return self.min(root.left)
    #Delete data
    def deleteDataHelper(self,root,data):
        if root==None:
            return False,None
        
        if root.data < data:
            deleted,newRight=self.deleteDataHelper(root.right,data)
            root.right=newRight
            return deleted,root
        
        if root.data > data:
            deleted,newLeft = self.deleteDataHelper(root.left,data)
            root.left=newLeft
            return deleted,root
        
        #root is leaf
        if root.left== None and root.right == None:
            return True,None
        
        #root has one child
        if root.left==None:
            return True,root.right
        if root.right==None:
            return True,root.left
        
        #root has both the children
        replacement = self.min(root.right)
        root.data=replacement
        deleted,newRightNode=self.deleteDataHelper(root.right,replacement)
        root.right=newRightNode
        return True,root
        
        
        
    
    def deleteData(self,data):
        deleted, newRoot = self.deleteDataHelper(self.root, data)
        if deleted:
            self.numNode -= 1
        self.root = newRoot
        return deleted 
    
            
        
    def count(self):
        return self.numNode
```

```python
b=BST()
b.insert(10)
b.insert(5)
b.insert(12)
print(b.isPresent(10))
print(b.isPresent(7))
print(b.count())
print(b.deleteData(4))
print(b.count())
b.printTree()
```
