```python
class GenericTreeNode:
    def __init__(self,data):
        self.data = data
        self.children= []
```

```python
import queue

def takeLevelwiseInput():
    q=queue.Queue()
    print("Enter root")
    rootData = int(input())
    
    if rootData == -1:
        return None
    root = GenericTreeNode(rootData)
    q.put(root)
    
    while not(q.empty()):
        curr_node = q.get()
        
        print("enter number of children for ",curr_node.data)
        numChildren = int(input())
        
        
        if numChildren >0:
            for i in range (numChildren):
                print("Enter next child of ",curr_node.data)
                childData = int(input())
                child = GenericTreeNode(childData)
                curr_node.children.append(child)
                q.put(child)
                
    return root
```

```python
def printDetailedTree(tree):
    if tree == None:
        return
    
    print(tree.data,":",end=" ")
    
    for child in tree.children:
        print(child.data,",",end=" ")
        
    print()
    
    for child in tree.children:
        printDetailedTree(child)
```

```python
import queue
def isIdenticalM2(tree1, tree2):

    if tree1 is None and tree2 is None:
        return True

    if (tree1 is not None and tree2 is None) or (tree1 is None and tree2 is not None):
        return False

    q1 = queue.Queue()
    q2 = queue.Queue()

    q1.put(tree1)
    q2.put(tree2)

    while not q1.empty() and not q2.empty():
        curr1 = q1.get()
        curr2 = q2.get()

        # Compare the data of the current nodes
        if curr1.data != curr2.data:
            return False

        # Enqueue child nodes
        children1 = curr1.children
        children2 = curr2.children

        if len(children1) != len(children2):
            return False

        for i in range(len(children1)):
            q1.put(children1[i])
            q2.put(children2[i])

    # If both queues are empty, the trees are structurally identical
    return q1.empty() and q2.empty()
```
