```python
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        

        
        
def treeInput():
    rootData=int(input())
    
    if rootData ==-1:
        return None
    
    
    root=BinaryTreeNode(rootData)
    
    leftTree = treeInput()
    rightTree=treeInput()
    root.left=leftTree
    root.right=rightTree
    
    return root        
        
def treeHeight(root):
    if root == None:
        return 0
    
    return 1+max(treeHeight(root.left),treeHeight(root.right))

def isBalanced(root):
    if root==None:
        return True
    
    lh=treeHeight(root.left)
    rh=treeHeight(root.right)
    
    if (lh-rh)>1 or (rh-lh)>1:
        return False
    
    isLeftBalanced = isBalanced(root.left)
    isRightBalanced = isBalanced(root.right)
    
    if isLeftBalanced and isRightBalanced:
        return True
    else:
        return False
    
def printTree(root):
    if root == None:
        return
    
    print(root.data,end=":")
    
    if root.left !=None:
        print("L ",root.left.data,end=",")
        
    if root.right !=None:
        print("R ",root.right.data,end=" ")
    print()
        
    printTree(root.left)
    printTree(root.right)
    
    

        
root=treeInput()
printTree(root)
print(isBalanced(root))
```
